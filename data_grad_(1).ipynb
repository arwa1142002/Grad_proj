```



# Creating 'Early Stages' and 'Advanced Stages' columns and initializing them with 'non'
df['Early Stages'] = 'non'
df['Advanced Stages'] = 'non'

# Conditions based on different types in 'Tumour Type 3' column
conditions = {
    'Combined hepatocelllular carcinoma and cholangiocarcinoma': {
        'Early Stages': ['Weight loss', 'Loss of appetite', 'Generalized weakness and fatigue'],
        'Advanced Stages': [
            'Abdominal pain or tenderness, especially in the upper right part',
            'Enlarged liver or palpable mass in the abdomen',
            'Jaundice',
            'Ascites',
            'Swelling in the abdomen or legs',
            'Easy bruising or bleeding'
        ]
    },
    'Hepatocellular carcinoma': {
        'Early Stages': ['Weight loss', 'Loss of appetite', 'Generalized weakness and fatigue'],
        'Advanced Stages': [
            'Abdominal pain or tenderness, especially in the upper right part',
            'Enlarged liver or palpable mass in the abdomen',
            'Jaundice',
            'Ascites',
            'Swelling in the abdomen or legs',
            'Easy bruising or bleeding'
        ]
    },
    'Hepatoblastoma': {
        'Early Stages': ['Abdominal swelling or mass', 'Weight loss', 'Loss of appetite', 'Pain or discomfort in the abdomen', 'Fever'],
        'Advanced Stages': ['Symptoms often vary depending on the age of the child and stage of cancer']
    },
    'Other liver': {
        'Early Stages': 'non',
        'Advanced Stages': 'non'
    },
    'Unspecified liver': {
        'Early Stages': 'non',
        'Advanced Stages': 'non'
    },
    'Ampulla of Vater': {
        'Early Stages': ['Often asymptomatic', 'Vague symptoms like indigestion or discomfort'],
        'Advanced Stages': [
            'Jaundice',
            'Abdominal pain, especially in the upper abdomen',
            'Unexplained weight loss',
            'Loss of appetite',
            'Nausea or vomiting',
            'Pale-colored stools and dark urine',
            'Itching'
        ]
    },
    'Cholangiocarcinoma': {
        'Early Stages': ['Often asymptomatic or mild symptoms', 'Abdominal pain or discomfort, especially in the upper right part', 'Unexplained weight loss'],
        'Advanced Stages': [
            'Jaundice',
            'Itching',
            'Dark urine and pale-colored stools',
            'Abdominal mass or swelling',
            'Fever, fatigue, and night sweats'
        ]
    },
    'Gallbladder': {
        'Early Stages': ['Often asymptomatic or mild symptoms', 'Abdominal pain, especially in the upper right part', 'Nausea and vomiting', 'Loss of appetite'],
        'Advanced Stages': [
            'Jaundice',
            'Unexplained weight loss',
            'Fever',
            'Abdominal mass or swelling',
            'Enlarged liver'
        ]
    }
}

# Applying conditions to fill 'Early Stages' and 'Advanced Stages' columns
for tumor_type, stages in conditions.items():
    df.loc[df['Tumour Type 3'] == tumor_type, 'Early Stages'] = ', '.join(stages['Early Stages']) if isinstance(stages['Early Stages'], list) else stages['Early Stages']
    df.loc[df['Tumour Type 3'] == tumor_type, 'Advanced Stages'] = ', '.join(stages['Advanced Stages']) if isinstance(stages['Advanced Stages'], list) else stages['Advanced Stages']

print(df)
```
